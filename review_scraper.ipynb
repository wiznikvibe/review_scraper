### Review Scraper 

Go to flipkart > search for product > Extract review and the rating associated with it 

Why this?

        - Can be used for building sentiment analysis and machine learning models 
        
        - Helps building data set from scratch, also for building data creation pipeline.
        
        - For NLP Projects, To understand the public opinion about certain products 

Steps- 

        -  Search for > Product 
        
        -  Html > View Page Source> Right Click on the browser 
        
        -  Look for the html tags with the required information and fetch them  using bs4 and requests library. 

```python
# Import Libraries
from bs4 import BeautifulSoup as bs 
from urllib.request import urlopen as urReq
import requests
import pandas as pd
```

```python
# Search for Products for data gathering 
flipkart_url= "https://www.flipkart.com/search?q="
```

```python
product = "oneplus11"
query = product.replace(" ","")
```

```python
# urReq returns the html page sorce for the url passed 
response_web = urReq(flipkart_url+query)
response_web
```

```python
# Make the Html page readable  
data = bs(response_web, 'html.parser')
data
```

```python
# find_all is used for finding the desired html objects and associated tags 
containers = data.find_all("div", {"class":"_1AtVbE col-12-12"})
len(containers)
```

```python
product3 = "https://www.flipkart.com"+containers[3].div.div.div.a['href']
product3
```

```python
product3_page = requests.get(product3)
```

```python
# enconding the url in order to pass it through a html parser for extracting reviews and ratings 
product3_page.encoding = 'utf-8'
```

```python
product_page3 = bs(product3_page.text,'html.parser')
product_page3
```

```python
# Listing all the reviews 
all_reviews = product_page3.find_all('div',{'class':"_16PBlm"})
len(all_reviews)
```

```python
# Review by Customers 
review = all_reviews[3].div.find_all('div', {'class':""})[0].div.text
review
```

```python
for i in range(len(all_reviews)-1):
    print( all_reviews[i].div.find_all('p',{'class':'_2sc7ZR'})[1].text)
```

```python
# Review By: Name 
name = all_reviews[5].div.div.find_all('p',{"class":"_2sc7ZR _2V5EHH"})[0].text
name
```

```python
# Review Header 
comment_header = all_reviews[5].find_all('p',{'class':"_2-N8zT"})[0].text
comment_header
```

```python
# Review 
review = all_reviews[5].div.find_all('div',{"class":""})[0].div.text
review
```

```python
# Date of Review Posted 
date = all_reviews[5].div.find_all('p',{'class':'_2sc7ZR'})[1].text
date
```

```python
# Price for Product
price = product_page3.find_all('div',{"class":"_30jeq3 _16Jk6d"})[0].text
price
```

```python
# Offer for the product 
offer = product_page3.find_all('div',{"class":"_1V_ZGU"})[0].text
offer
```

```python
# Product Highlights 
highlight = product_page3.find_all('div',{"class":"_2418kt"})[0].text
highlight
```

```python
# 2:12:17 
comment_headers = []
ratings = []
customer_names = []
reviews =  []
dates = []
prices = []
highlights = []
for i in range(len(all_reviews)-1):
    comment_headers.append(all_reviews[i].find_all('p',{'class':"_2-N8zT"})[0].text)
    reviews.append(all_reviews[i].div.find_all('div',{"class":""})[0].div.text)
    ratings.append(all_reviews[i].text[0])
    customer_names.append(all_reviews[i].div.div.find_all('p',{"class":"_2sc7ZR _2V5EHH"})[0].text)
    dates.append(all_reviews[i].div.find_all('p',{'class':'_2sc7ZR'})[1].text)
    prices.append(product_page3.find_all('div',{"class":"_30jeq3 _16Jk6d"})[0].text)
    highlights.append(product_page3.find_all('div',{"class":"_2418kt"})[0].text)
    

```

```python
dict_data = {"date":dates,
            "customer_name":customer_names,
            
            "Comment":comment_headers,
             "price":prices,
             "highlight":highlights,
             "review":reviews,
             
             "rating":ratings,
            }

review_df = pd.DataFrame(dict_data)
review_df.head() 
```
