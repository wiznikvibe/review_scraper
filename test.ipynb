```python
# 21:07
from bs4 import BeautifulSoup as bs 
from urllib.request import urlopen as urReq
import requests
```

```python
flipkart_url= "https://www.flipkart.com/search?q="+"nothing+phone+1"
```

```python
#https://www.flipkart.com/search?q=iphone12
response_web = urReq(flipkart_url)
response_web
```

```python
data = bs(response_web, 'html.parser')
data
```

```python
containers = data.find_all("div", {"class":"_1AtVbE col-12-12"})
```

```python
product3 = "https://www.flipkart.com" + containers[3].div.div.div.a['href']
```

```python
product_3 = requests.get(product3)
```

```python
product3
```

```python
product_3.encoding = 'utf-8'
```

```python
product_3_page = bs(product_3.text,'html.parser')
product_3_page 
```

```python
all_review =  product_3_page.find_all('div',{'class':"_16PBlm"})
```
